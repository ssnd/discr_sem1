

### Формальное определение функции алгебры логики
*Определение.* Пусть $U = \{u_1, u_2, ..., u_m, ...\}$ - исходный алфавит перменных.
Будем рассматривать функции $f(u_{i_1}, u_{i_2},..., u_{i_n}) (u_{i_\nu} \neq u_{i_\mu}$ при  $\nu\neq\mu$), аргументы которых определены на множестве $E_2=\{ 0,1\}$ и такие, что  $f(\alpha_1, \alpha_2, ..., \alpha_n)\in E_2$, когда $\alpha_i\in E_2\ (i=1,2,...,n)$  
Эти функции будем называть **функциями алгебры логики**  или **булевыми функциями**.
Из определения следует, что для задания такой функции достаточно указать, какое значение функции соответствует каждому из наборов значений аргументов, т.е. выписать таблицу истинности.



### Существенная зависимость от аргумента. Фиктивные переменные
*Определение.* Функция $f(x_1, ..., x_{i-1}, x_i, x_{i+1}, ..., x_n)$ из $P_2$ **зависит существенным образом** от аргумента $x_i$, если существуют такие значения $\alpha_1, ..., \alpha_{i-1}, \alpha_{i+1}, ..., \alpha_n$ переменных $x_1,...,x_{i-1}, x_{i+1}, ..., x_n$, что   
$f(\alpha_1, ..., \alpha_{i-1}, 0, \alpha_{i+1}, ..., \alpha_n) \neq f(\alpha_1, ..., \alpha_{i-1}, 1, \alpha_{i+1}, ..., \alpha_n)$  
В этом случае перменная $x_i$ называется __существенной__. Если $x_i$ не является существенной переменной, она **фиктивная** или **несущественная**.



### Операции введения и удаления фиктивной перменной
Пусть для функции $f(x_1, ..., x_n)$ переменная $x_i$ является фиктивной. Возьмём  таблицу для функции $f(x_1, ..., x_n)$ и по ней построим новую таблицу путём вычёркивания всех строк вида $(\alpha_1, ..., \alpha_{i-1}, 1, \alpha_{i+1}, ..., \alpha_n)$ и вычёркивания столбца для аргумента $x_i$. Полученная таблица будет определять некоторую функцию  $g(x_1,...,x_{i-1}, x_{i+1},..., x_n)$. Будем говорить, что функция $g$ получена из  $f$ путём _удаления фиктивной переменной $x_i$, а также, что функция $f$  получается из $g$  путём введения фиктивной переменной._

### Понятия равенства булевых функций
_Определение._  Функции $f_1$ и  $f_2$ называются **равными**, если функцию $f_2$ можно получить из $f_1$  путём добавления или изъятия фиктивных переменных.
### Определение формулы и подформулы

_Определение._ Пусть $B$ - некторое (не обязательно конечное) подмножества функций из $P_2$.  
a) _Базис индукции_. Каждая функция $f(x_1, ..., x_m)$ из $B$ называется _формулой над $B$_.  
б) _Индуктивный переход_. Пусть $f_0(x_1,...,x_m)$  --  функция из $B$ и $A_1, ..., A_m$ -- выражения, являющиеся либо формулами над $B$ либо символами переменных из $U$.  
Тогда выражение $f_0(A_1,.., A_m)$   называется **формулой над $B$**.

Пусть $D$ - произвольная формула над $B$, тогда формулы, которые использовались для её построения будем называть __подформулами__ формулы $D$

### Сопоставление формуле булевой функции - формальное задание.
Сопоставим каждой формуле $D(x_1, ..., x_n)$ над $B$ функцию $f(x_1, ..., x_n)$ из $P_2$ опираясь на индуктивное определение формул.  
а) _Базис индукции_. Если $D(x_1,...,x_n) = f(x_1,...,x_n)$, где $f\in B$, то формуле $D(x_1,..., x_n)$ сопоставим функцию $f(x_1,...,x_n)$  
б) _Индуктивный переход_. Пусть $D(x_1, ..., x_n) = f_0(A_1,..., A_m)$, где $A_i (i=1,...,m)$ является либо формулой над B, либо символом переменной $x_{j(i)}$.  
Тогда по предположению индукции $A_i$ сопоставлена либо функция $f_i$ из $P_2$ либо тождественная функция $f_i=x_{j(i)}$.
Сопоставим формуле $D(x_1,..., x_n)$ функцию $f(x_1,...,x_n)=f_0(f_1,...,f_m)$
### Совершенная КНФ и совершенная ДНФ

$x^\sigma = x\sigma \lor \overline{x}\overline{\sigma}$
Где $\sigma$ - параметр, равный либо 0, либо 1. $x^\sigma =1 \Leftrightarrow x=\sigma$

Теорема 3. Каждую функцию алгебры логики $f(x_1,...,x_nn)$ при любом $m (1\leq m\leq n)$ можно представить в следующей форме:
$f(x_1,..., x_m, x_{m+1}, ..., x_n) = \lor_{(\sigma_1,...\sigma_m)}\  x_1^{\sigma_1} \&\ ...\ \&\ x_m^{\sigma_m}\ \&\ f(\sigma_1,\ ...,\  \sigma_m, x_{m+1}, ..., x_n)$ где дизъюнкция берётся по всевозможным наборам значений переменных $x_1,...,x_m$.
__Доказательство__. Рассмотри произвольный набор значений переменных $(\alpha_1,.., \alpha_n)$ и покажем, что левая и правая части предыдущего соотношения  принимают на нём одно и то же значение.
Левая часть даёт $f(\alpha_1,...,\alpha_n)$. Правая -   
$\lor_{(\sigma_1,...,\sigma_m)} \alpha_1^{\sigma_1}\ \&\ ...\ \&\ \alpha_m^{(\sigma_m)}\ \&\ f(\sigma_1,...,\sigma_m,\alpha_{m+1}, ..., \alpha_n) = $

$\alpha_1^{\alpha_1}\ \&\ ...\ \&\ \alpha_m^{\alpha_m}\&f(\alpha_1, ... ,\alpha_m, \alpha_{m+1}, ..., \alpha_n) = f(\alpha_1,...,\alpha_n)$

Одно из следствий - разложение по всем переменным:
$f(x_1,...,x_n) = \lor_{(\sigma_1,...,\sigma_n)} x_1^{\sigma_1} \&\ ...\ \&\ x_n^{\sigma_n}\ \&\ f(\sigma_1,...,\sigma_n)$  
При $f(x_1,...,x_n)\neq 0$ оно может быть преобразовано: 
$\lor_{(\sigma_1,...,\sigma_n)} x_1^{\sigma_1}\ \&\ ...\ \&\ x_n^{\sigma_n}\ \&\ f(\sigma_1, ..., \sigma_n) =$
$\lor_{(\sigma_1,...,\sigma_n)} x_1^{\sigma_1}\ \&\ ...\ \&\ x_n^{\sigma_n}$

Такое разложение носит название __совершенной дизъюнктивной нормальной формы__ (совершенной днф)


Совершенная ДНФ есть выражение типа $\sum\prod$ т.е. логическая сумма произведений. Но можно получить и разложение типа $\prod\sum$.

Если $f\not\equiv 1$ разложим функцию $f*$ в СДНФ:
$f*(x_1,...,x_n) = \lor_{(\sigma_1,...,\sigma_n)} x_1^{\sigma_1}\ \&\ ...\ \&\ x_n^{\sigma_n}$

Т.к. $f=f^{**}$,  $f^{**}(x_1,...,x_n)=\&_{(\sigma_1,...,\sigma_n)} (x_1^{\sigma_1}\lor ...\lor x_n^{\sigma_n})$, это выполняется при $f^*(\sigma_1, ...\sigma_n)=1 \Leftrightarrow f(\overline{\sigma_1}, ..., \overline{\sigma_n})=0 \Rightarrow f=\&_{(\sigma_1,...,\sigma_n)} (x_1^{\overline{\sigma_1}}\lor ...\lor x_n^{\overline{\sigma_n}})$ при $f(\sigma_1,...,\sigma_n)=0$ - совершенная конъюнктивная нормальная форма.

### Полином Жегалкина
Теорема: каждая функция алгебры логики может быть выражена в виде формулы через отрицание, конъюнкцию и дизъюнкцию.  
Доказательство: Если $f\equiv 0$ тогда $f(x_1,...,x_n)=x_1\&\overline{x_1}$ в другом случае представляем её сднф. Доказано.  
Определение: Система функций из $P_2$ называется полной, если любая булева функция может быть записана в виде формулы через функции этой системы.  
Пример $B=\{\overline{x}, x_1\&x_2, x_1\lor x_2\}$ - полная система.  
Система $B={0,1, x_1 x_2, x_1+x_2}$ - полная. $(x_1+1=\overline{x_1}, x_1x_2=x_1\&x_2)$  
Теорема. Каждая функция из $P_2$ может быть выражена при помощи полинома по $\mod 2$ (Полинома Жегалкина).
Подсчитаем число полиномов Жегалкина от переменных  $x_i, ..., x_n$, т.е. число выражений вида  
$\sum_{(i_1,...,i_s)} a_{i_1...i_s}x_{i_1}...x_{i_s}$  
Число членов $x_{i_1}...x_{i_s}$ равно количеству подмножеств $(i_1,...,i_s)$ из $n$  чисел $(1,...,n) $ т.е. $2^n$. Поскольку  $a_{i_1...i_s} $ равно 0 или 1, искомое число полномов равно $2^{2^n}$ т.е. числу всех булевых функций от тех же переменных. Отсюда, как следствие, получаем единственность представления функция посредством полиномов Жегалкина.

### Определение предполных классов $P_0, P_1, S, L, M$
**$P_0$ - класс булевых функций, сохраняющих константу 0**, т.е. функций, для которых выполнено равенство $f(0,...,0)=0$  
**$P_1$ - класс булевых функций, сохраняющих константу 1**, т.е. функций, для которых выполнено равенство $f(1,...,1)=1$  
**$S$ - класс самодвойственных функций, таких что $f^*=f$**, т.е. функций для которых имеет место тождество $\overline{f}(\overline{x}_1, ..., 
\overline{x}_n) = f(x_1,...,x_n)$

**Определение.** Для двух наборов $\alpha=(\alpha_1,...,\alpha_n)$ и $(\beta=\beta_1,...,\beta_n)$ выполнено отношение предшествования $\alpha\prec\beta$ б если $\alpha_1\leq\beta_1,..., \alpha_n\leq\beta_n$.   
Функция называется **монотонной**, если для любых двух наборов $\alpha$ и $\beta$  таких что $\alpha\prec\beta$ имеет место неравенство $f(\alpha)\leq f(\beta)$   
**М - множество всех монотонных функций **  
**L - класс всех линейных функций.** Он содержит константы 0,1, $x$, $\overline{x}$ и $x_1+x_2$

### Формулировка Критерия полноты Поста
**Критерий Поста** Для того, чтобы система функций была полной, необходимо и достаточно, чтобы она целиком не содержалась ни в одном из пяти замкнутых классов $P_0, P_1, S, M, L$


### Базовые определения из теории графов: понятие графа, мультиграфа, орграфа, псевдографа.
Граф - пара множеств $(V,E)$ где $E$- множество пар элементов $V$. Элементы множеств $V$ и $E$ называют вершинами и рёбрами графа соответственно.  
Если в E есть повторяющиеся пары - мультиграф, а сами пары - рёбра/дуги.  
Если в E есть пары, оба элемента которых совпадают, то говорят о псевдографе, а сами пары - петли  
### Определение пути, цикла, клики, независимого множества в графах
Маршрут - последовательность вершин и рёбер графа, в которой последовательные элементы инцидентны друг другу.   
Путь - незамкнутый маршрут без повторений рёбер.  
Цикл - замкнутый маршрут (начало совпадает с концом) без повторяющихся рёбер.  
Полный граф - граф,  в  к отором есть все возможные рёбра.  
Клика - полный подграф.  
Независимое множество - подмножество вершин, порождающее пустой подграф  (есть множество, такое, что любые две вершины в нем не смежны, никакая пара вершин не соединена ребром).  
### Определение числа независимости, кликового числа графа. Связность графа. Подграфы и порождённые подграфы.
Число независимости - размер максимального независимого множества  
Кликовое число - максимальный размер клики в графе.  
Связный граф - граф, в котором между любыми двумя вершинами существует путь.  
Компонента связности графа - максимальный связный подграф.  
Мост - ребро, удаление которого приводит к графу с бОльшим числом компонент связности.  
Точка сочленения - вершина, удаление которой приводит к графу с бОльшим числом компонент связности.  

### Подграфы и порождённые подграфы
Пусть $G = (V, E)$ и $G'=(V', E')$ - графы.  
$G'$ является подграфом графа $G$, если $V'\subseteq V$ и  $E' \subseteq E$  
$G'$ - остовный подграф, если $V'=V$  
$G'$ - порождённый подграф графа $G$, если с каждой парой вершин он содержит и все инцидентныеим рёбра: $\forall u,v\ (uv\in E\ и\ u,v\in V')\Rightarrow uv\in E'$
### Шесть определений дерева.
Эквивалентные определения дерева:  
1) Связный граф без циклов  
2) Связный граф, при удалении любого ребра становящийся несвязным  
3) Граф без циклов, в котором при добавлении любого ребра появляется   цикл  
4) Граф, в котором между любой парой вершин сущесвтует единственный  путь   
5) Связный граф, в котором #вершин = 1 + #рёбер  
6) Граф без циклов, в котором  #вершин = 1 + #рёбер  

### Эйлеров граф. Критерий эйлеровости графа.
Эйлеров цикл - цикл, проходящий через каждое ребро графа ровно по одному разу.
Эйлеров граф - граф, в котором есть эйлеров цикл.



### Определение гамильтонова цикла и гамильтонова графа.
Гамильтонов цикл - через каждую вершину графа проходит ровно по одному разу.
Гамильтонов граф - граф, в котором есть гамильтонов цикл.

### Признак Дирака и признак Оре. Признак Эрдеша-Хватала.
**Теорема Оре**  
Если для любых $u,v\in V(G)$ выполнено $\deg u + \deg v\geq|V(G)|$, то G - Гамильтонов.    
**Теорема Дирака.**  
Если $G$ - неорьентированный граф и $\delta$  - минимальная степень его вершин. Если $n\geq 3$ и $\delta\geq n/2$,  то $G$ - гамильтонов.  

Пусть дан граф $G=(V, E)$, тогда его числом независимости называется число  $\alpha(G)=\max\{k\in \mathbb{N}: \exists W\subseteq V: |W|=k\ \&\ \forall x,y\in W (x,y)\not\in E\}$
Вершинной связностью графа, обозначаемой $k(G)$ называется минимальное количество вершин, в результате удаления которых граф перестаёт быть связным.

**Признак Эрдеша-Хватала**
Пусть $G=(V,E)$ - граф, такой что $|V|\geq 3$ и $\alpha(G)\leq k(G)$. Тогда $G$ - гамильтонов.

### Понятие укладки графов. Понятие планарного графа. Формула Эйлера, оценки для числа рёбер планарного связного графа. Критерий Понтрягина-Куратовского и Вагнера.

Укладка графа - сопоставление:  
- вершинам графа - точек поверхности
- рёбрам графа - гладких кривых без самопересечений  
так, чтобы
- для любого ребра $xy$ концы соответствующей кривой совпадали с точками, сопоставленными вершинам $x$ и $y$
- кривые, соответсвующие рёбрам, не пересекались (за исключением, быть может, своих концов).


Планарный граф - граф, для которого существует плоская укладка.

Формула Эйлера.
Для любого связного планарного графа 
\#вершин - \#рёбер+\#граней = 2

Для любого планарного графа, в котром минимальная длина циклов равна $t$ выполнено неравенство $|V| \geq\frac{t}{2}\cdot |E| $


Два графа гомеоморфны, если один из них может быть получен из другого применением конечного числа операций добавления и исключения проходных вершин и склеивания ребер.


Теорема Понтрягина — Куратовского: Граф G – планарен $\Leftrightarrow$ G не содержит подграфов гемеоморфных графам $K_5$ или $K_{3,3}$.
Теорема Вагнера: Граф G – планарен $\Leftrightarrow$ миноры G не содержат $K_5$ или $K_{3,3}$.


### Формальное определение машины Тьюринга
Машиной Тьюринга называется кортеж $(\sum, \Gamma, Q, q_1, q_0, \delta)$, где $\sum,\ \Gamma\ и\ Q$ суть конечные непустые множества, причём $\sum\subset\Gamma$ и $\Gamma \cap Q=\varnothing, q_0, q_1\in Q, q_0\neq q_1$, а $\delta :(Q\setminus\{q_0\})\times \Gamma \rightarrow Q\times\Gamma\times\{L,N,R\}$. $\sum$  называется входным алфавитом, $\Gamma$ - ленточным алфавитом, $Q$ - множеством состояний, а $q_1, q_0$  - начальным и завершающим состояниями соответственно. Функцию перехода $\delta$ записывают в форме списка команд вида $q_ia_j\rightarrow q_ra_sD$

Среди элементов Г выделяют специальный символ #, не входящий в множеств $\sum$.
Если К - конечное множество, то через К* обозначается множество всех конечных последовательностей элементов из этого мноежства (слова в алфавите К). Последовательность из нуля элементов обозначается $\varepsilon$ и называется поустым словом. На последовательностях определена операция конкатенации (припипсывания обозначаемая символом $\cdot$)


### Определение вычислимой функции. Понятие о разрешимом и перечислимом множестве. Теорема о множественных операциях (объединение, пересечение, дополнение, проекция), сохраняющих перечислимость и разрешимость. Критерий поста о разрешимости множества. 
Функция $f:A\rightarrow B $  вычислима, если существует программа, которая на любом входе $x\in dom f$ выписывает $f(x)$ и завершается, а на любом входе $x\in A\setminus dom f$  не завершается ни за какое конечное количество шагов.



### Понятие о разрешимом и перечислимом множестве.

Множество А разрешимо, если его характеристическая функция $\chi_A(x)$ вычислима.  
Множество А перечислимо, если есть программа, на пустом входе последовательно выписывающая все элемента А и только их.  


### Теорема о множественных операциях (объединение, пересечение, дополнение, проекция), сохраняющих перечислимость и разрешимость.

Лемма. 
1. Если А - разрешимо, оно перечислимо.
2. Если А, B разрешимы, то разрешимы и $A\cup B, A\cap B, A\times B, \overline{A}$;
3. Если А, В перечислимы, то перечислиым  $A\cup B, A\cap B, A\times B, pr^iA$;

Понятно, что любое разрешимое множество перечислимо. Для построения алгоритма его перечисления можно в цикле вызывать характер. функцию  для А, выдавая в память (перечисляя) только числа для которых она равна 1.

Покажем, например, что множество 𝐴 × 𝐵 перечислимо, если перечислимы 𝐴 и 𝐵. Станем попеременно выполнять по шагу перечисляющих алгоритмов для 𝐴 и для 𝐵, раздельно выписывая куда-либо всякий получаемый элемент каждого из этих множеств. Как только получен очередной элемент 𝑎 ∈ 𝐴, выдадим на выход пары (𝑎,𝑏𝑖) для всех накопленных к этому моменту элементов 𝑏1, . . . , 𝑏𝑘 ∈ 𝐵. Аналогично поступим с очередным элементом множества 𝐵. Если (𝑎, 𝑏) ∈ 𝐴 × 𝐵, то элементы 𝑎 и 𝑏 попадут в наш «накопитель», причем один из них попадет туда поз- же другого. Пусть, без ограничения общности, это будет 𝑎. Тогда 𝑏 уже содержится в накопителе, и мы выпишем пару (𝑎, 𝑏). Также очевидно, что никаких лишних пар мы не выписываем.



### Критерий поста о разрешимости множества.
Критерий Поста разрешимости множества. Множество А разрешимо тогда и только тогда когда $A$ и $\overline{A}$ перечислимы.

Импликация слева направо следует из пред. леммы. Допустим что эти два множества перечислимы. Как мог бы работать разрешающий алгоритм А? ПОлучив на вход какое=то число достаточно попеременно выполнять по шагу перечисляющих алгоритмов для обоих множест, ожидая, пока выбранное число не будет выписано тем или другим. Поскольку оно принадлежит одному из множеств, процедура завершится.


### Альтернативные определения перечислимых множеств

Пусть $A\subseteq\mathbb{N}$. Тогда равносильны условия
1. А перечислимо.
2. существует вычислимая $f:\mathbb{N}\overset{p}{\rightarrow}\mathbb{N}$ т.ч. $A=dom\ f$
3. существует вычислимая $f:\mathbb{B}\overset{p}{\rightarrow}\mathbb{N}$ т.ч. $A=rng\ f$
4. $A=\emptyset$ или существует вычислимая тотальная $f:\mathbb{N}\rightarrow\mathbb{N}$ т.ч. $A = rng\ f$
5. существует разрешимое $B\subseteq\mathbb{N}^2$ т.ч. $A=pr^1B$ 

### Универсальная вычислимая функция. Существование увф. Теорема о том, что объединение двух перечислимых множеств можно представить как объединение двух непересекающихся перечислимых множеств.
Функция  $U: \mathbb{N}^2\overset{p}{\rightarrow}\mathbb{N}$называется универсальной вычислимой, если она вычислима и для всякой вычислимой функции $f:\mathbb{N}\overset{p}{\rightarrow}\mathbb{N}$ найдётся число  $n\in\mathbb{N}$ называемое индексом функции $f$ относительно $U$, т.ч.  $U_n=f$,  т.е.   
$\forall x(f(x)\backsimeq U(n,x))$  
Т.О. U нумерует все вычислимые функции и только их: отображение $n\rightarrow U_n$  является тотальной сюръекцией из $\mathbb{N}$ в класс вычислимых функций одного аргумента. Поэтому универсальные вычислимые функции мы также будем называть нумерациями. 

Существование. Универсальную вычислимую функцию 𝑈 будет вычислять программа-«интерпретатор», получающая номер 𝑛 и входное значение 𝑥, вычисляющая текст программы Π(𝑛), и начинающая шаг за шагом исполнять полученную программу на входе 𝑥.

Можно утверждать существование такой программы, сославшись на тезис Тьюринга-Чёрча, или непосредственно предъявить ее для нашей модели вычислений. Скажем, указать универсальную машину Тьюринга 𝒰 , которая принимает на вход два натуральных числа (т. е. их коды) 𝑛 и 𝑥. После этого 𝒰 запускает машину Тьюринга, вычисляющую нумерацию Π, на входе 𝑛, которая перерабатывает этот вход в как-либо закодированную програм- му машины Тьюринга 𝑀𝑛. Затем машина 𝒰 последовательно применяет инструкции машины 𝑀𝑛 ко входу 𝑥. Если машина 𝑀𝑛 должна перейти в заключительное состояние, машина 𝒰 стирает с ленты все, кроме результата работы 𝑀𝑛, и тоже переходит в заключительное состояние. Если 𝑀𝑛 продолжает работать неопределенно долго, универсальная машина делает это вместе с нею.


### Т-предикаты и их вычислимость. 
Для каждой УВФ $U$ разрешимы множества   
$T'=\{(n,x,y,k)\ |\ $  алгоритм останавливается на входе $(n,x)$ ровно за $k$ шагов и оставляет на выходе $y\}$  и   
$T'=\{(n,x,k)\ |\ $  алгоритм останавливается на входе $(n,x)$ ровно за $k$ шагов $\}$

неформально, нужно просто остановить алгоритм на входе после к шагов и посмотреть, завершился ли он , и, если завершился, что подал на выход. ИЛи констатировать, что алгоритм остановился раньше, чем за к шагов.

Покажем, что произвольные перечислимые множества $A,B\subseteq \mathbb{N}$ можно разделить на перечислимые непересекающиеся части. Именно, существуют перечислимые $A',B'$ такие, что $A'\subseteq A, B'\subseteq B, A'\cup B'=A\cup B,\ A'\cap B'=\emptyset$

Проблема в том, как поделить между множествами 𝐴 и 𝐵 их пересече- ние. Отдать все одному из множеств не всегда возможно: 𝐴 ∖ 𝐵 не обязано быть перечислимым. Однако можно, например, посмотреть, в перечисле- нии какого из множеств элемент 𝑛 ∈ 𝐴 ∩ 𝐵 встретится на шаге с меньшим номером — тому множеству его и отдать. При этом все элементы, скажем, 𝐴 ∖ 𝐵 заведомо будут «раньше» выписаны в перечислении 𝐴.

### Главная универсальная вычислимая. Существование ГУВФ. Любая ГУВФ является также УВФ.

Вычислимая функция $U:\mathbb{N}^2\overset{p}{\rightarrow}\mathbb{N}$ называется главной, если для любой вычислимой функции $V:\mathbb{N}\overset{p}{\rightarrow}\mathbb{N}$ найдётся вычислимая тотальная функция $s:\mathbb{N}\rightarrow\mathbb{N}$ такая что $V_n=\{s(n)\}$ для всех $n\in\mathbb{N}$ т.е.  
$\forall x\forall n V(n,x)\backsimeq U(s(n),x)$


Лемма. Если $U$  гувф, то $U$  - увф.

Вычислимость U уже есть. Проверим универсальность. Пусть функция $f:\mathbb{N}\overset{p}{\rightarrow}\mathbb{N}$ вычислима. Рассмотрим функцию $V:\mathbb{n}^2\overset{p}{\rightarrow}\mathbb{N}$ т.ч. $V(n,x)\backsimeq f(x)$ для всех $n,x\in\mathbb{N}$. Очевидно, что $V$ вычислима. Как видим, все сечения $V_n$ фунцкии $V$ совпадают с $f$. В силу главности, найдётся вычислимая функция $s:\mathbb{N}\rightarrow\mathbb{N} $ т.ч. $U_{s(n)} = V_n=f$ при всех $n\in\mathbb{N}$. Поэтому $f$ будет иметь, например, индекс $s(0)$ относительно $U$


### Теорема о неподвижной точке.

Пусть универсальная вычислимая функция $U$ - главная, и вычислимая функция $f:\mathbb{N}\rightarrow\mathbb{N}$ тотальна. Тогда существует $n\in\mathbb{N}$, т.ч. $\{n\}=\{f(n)\}$ т.е. $\forall x (U(n,x))\backsimeq U(f(n),x))$

### Лемма о рекурсии.  Пример применения теоремы о неподвижной токе. 

Пусть функция $V:\mathbb{N}^2\overset{p}{\rightarrow}\mathbb{N}$ вычислима. Тогда существует $n$ т.ч. $U(n,x)\backsimeq V(n,x)$ для всех $x$. Иными словами, $\{n\}=V_n$.

### Лемма о совместной рекурсии. 

Пусть функции $V_1,V_2:\mathbb{N}^3\overset{p}{\rightarrow}\mathbb{N}$ вычислимы. Тогда существует $a,b\in\mathbb{N}$, т.ч. для всех $x\in\mathbb{N}$ выполнены  
$U(a,x)\backsimeq V_1(a,b,x)\ \ \&\ \ U(b,x)\backsimeq V_2(a,b,x)$

### Теорема Райса-Успенского 
Пусть нумерация U главная и множество F вычислимых функций одного аргумента нетривиально, т.е. найдётся $f\in F$ и найдётся вычислимая $g\not\in F$. Тогда множество индексов 
$F'=\{n|\{n\}\in F\}$ неразрешимо.


### m-сводимые множества. Свойства m-сводимых множеств.
Пусть $A,B\subseteq\mathbb{N}$. Множество A m-сводится к множеству В ($A\leq_mB$), если существует вычислимая тотальная функция $f:\mathbb{N}\rightarrow\mathbb{N}$ т.ч.  
$\forall x\in\mathbb{N}(x\in A \leftrightarrow f(x)\in B)$

Пусть $A,B,C\subseteq\mathbb{N}$. Тогда  
1. $A\leq_m A$  
2. если  $A\leq_m B$ и  $B\leq_m C$, то  $A\leq_m C$  
3. если  $A\leq_m B$, то $\overline{A}\leq_m \overline{B}$  
4. если  $A\leq_m B$ и В разрешимо, то А разрешимо.  
5. если  $A\leq_m B$ и В перечислимо, то А перечислимо.
