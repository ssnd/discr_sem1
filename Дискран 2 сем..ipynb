### –°–æ–ø–æ—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —Ñ–æ—Ä–º—É–ª–µ –±—É–ª–µ–≤–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ - —Ñ–æ—Ä–º–∞–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ.

### –°–æ–≤–µ—Ä—à–µ–Ω–Ω–∞—è –ö–ù–§ –∏ —Å–æ–≤–µ—Ä—à–µ–Ω–Ω–∞—è –î–ù–§

–¢–µ–æ—Ä–µ–º–∞ 3. –ö–∞–∂–¥—É—é —Ñ—É–Ω–∫—Ü–∏—é –∞–ª–≥–µ–±—Ä—ã –ª–æ–≥–∏–∫–∏ $f(x_1,...,x_nn)$ –ø—Ä–∏ –ª—é–±–æ–º $m (1\leq m\leq n)$ –º–æ–∂–Ω–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –≤ —Å–ª–µ–¥—É—é—â–µ–π —Ñ–æ—Ä–º–µ:
$f(x_1,..., x_m, x_{m+1}, ..., x_n) = \lor_{(\sigma_1,...\sigma_m)}\  x_1^{\sigma_1} \&\ ...\ \&\ x_m^{\sigma_m}\ \&\ f(\sigma_1,\ ...,\  \sigma_m, x_{m+1}, ..., x_n)$ –≥–¥–µ –¥–∏–∑—ä—é–Ω–∫—Ü–∏—è –±–µ—Ä—ë—Ç—Å—è –ø–æ –≤—Å–µ–≤–æ–∑–º–æ–∂–Ω—ã–º –Ω–∞–±–æ—Ä–∞–º –∑–Ω–∞—á–µ–Ω–∏–π –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö $x_1,...,x_m$.
__–î–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–æ__. –†–∞—Å—Å–º–æ—Ç—Ä–∏ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–π –Ω–∞–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö $(\alpha_1,.., \alpha_n)$ –∏ –ø–æ–∫–∞–∂–µ–º, —á—Ç–æ –ª–µ–≤–∞—è –∏ –ø—Ä–∞–≤–∞—è —á–∞—Å—Ç–∏ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è  –ø—Ä–∏–Ω–∏–º–∞—é—Ç –Ω–∞ –Ω—ë–º –æ–¥–Ω–æ –∏ —Ç–æ –∂–µ –∑–Ω–∞—á–µ–Ω–∏–µ.

$\alpha_1^{\alpha_1}\ \&\ ...\ \&\ \alpha_m^{\alpha_m}\&f(\alpha_1, ... ,\alpha_m, \alpha_{m+1}, ..., \alpha_n) = f(\alpha_1,...,\alpha_n)$
